it's clear that consistent upload is necessary for a successful channel. let's check the most frequent words in the titles

```python
all_title = df['title'].str.cat(sep='__')
```

```python
all_title
```

```python
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
all_title = all_title.lower()
all_title = ''.join(c for c in all_title if c.isalnum() or c.isspace())

words = word_tokenize(all_title)
english_stopwords = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in english_stopwords]

filtered_text = ' '.join(filtered_words)
filtered_text_list = filtered_text.split(' ')

freq_words = Counter(filtered_text_list).most_common()
```

```python
freq_words
```

lets make a wordcloud for this for easier interpretability

```python
from wordcloud import WordCloud
```

```python
freq_dict = {}
for key, val in freq_words:
    freq_dict[key] = val
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dict)

fig3, ax3 = plt.subplots(figsize=(10, 6))
ax3.imshow(wordcloud, interpolation='bilinear')
ax3.axis('off')

plt.show()
```
